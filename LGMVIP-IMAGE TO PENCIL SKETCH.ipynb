## IMPORTING CV2 TO CARRY OUT OPERATIONS 

```python
 import cv2
```

## IMPORTING THE IMAGE 

```python
image = cv2.imread('Tweety.jpg')
```

## VISUALIZING THE IMAGE 

```python
import matplotlib.pyplot as plt
image = cv2.imread('Tweety.jpg')
```

## USING [...,::-1] TO CONVERT THE IMAGE TO RGB FORMAT 

```python
image1 = plt.imshow(image[...,::-1])
image1
```

## CONVERTING IMAGE TO GRAYSCALE 

```python
image_to_grayscale = cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(image_to_grayscale,cmap='gray')
```

## INVERTING THE GRAY SCALE IMAGE 

```python
image2 = 255-image_to_grayscale
```

```python
plt.imshow(image2,cmap='gray')
```

## BLURRING THE GRAYSCALE IMAGE 

```python
image3 = cv2.blur(image2,(100,100),cv2.BORDER_DEFAULT)
```

```python
plt.imshow(image3,cmap='gray')
```

```python
image3 = cv2.GaussianBlur(image2,(5,5),0)
```

```python
plt.imshow(image2,cmap='gray')
```

## FINAL IMAGE

```python
image4 = 255-image3
```

```python
img = cv2.divide(image_to_grayscale,image4)
```

```python
plt.imshow(img,cmap='gray')
```
